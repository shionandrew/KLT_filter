```python
F_evalues, P_dagger, P_dagger_inv = [], [], []
for pp in range(2):
    evalues, P = la.eigh(F[pp], S[pp])
    F_evalues.append(evalues)
    P_dagger.append(P.T.conj())
    del P
    P_dagger_inv.append(la.inv(P_dagger[pp]))
[F_evalues[pp].shape for pp in range(2)]
```

```python
def klt_filter(visibility, ra, u, v, w, target_ra, target_dec, lat, fb_rfi, H, N_good_inputs):
    """
    """
    # Calculate hour angle
    ha = np.radians(ra - target_ra)[np.newaxis, np.newaxis, :]

    vis_out = []
    for pp in range(2):
        vis_int = visibility[pp].copy()
        # Construct fringestop phase
        fs_phase = cal_tools.fringestop_phase(ha, np.radians(lat), np.radians(target_dec), u[pp], v[pp], w)
        for fb in fb_rfi:
            Vfs = cal_tools.utvec2mat(N_good_inputs[pp], vis_int[fb, :, 0] * fs_phase[fb, :, 0])
            Vfs_klt = np.linalg.multi_dot([H[pp], Vfs, H[pp].conj().T])
            vis_int[fb, :, 0] = fs_phase[fb, :, 0].conj() * cal_tools.mat2utvec(Vfs_klt)
        vis_out.append(vis_int)
        
    return vis_out
```
