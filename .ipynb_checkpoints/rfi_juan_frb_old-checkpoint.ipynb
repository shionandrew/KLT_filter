```python
hf = figure(num=1, figsize=(20, 10))
matplotlib.rcParams.update({'font.size': 22})

time_index_frb_ds = 3000#.175/(downsample_factor*2.56e-6)
xlims = [2000,4000]#int(.100/(downsample_factor*2.56e-6)),int(.200/(downsample_factor*2.56e-6))]#, 240]
ylims = [-2, 5]
XLIM=False
YLIM=True
AV=False
f_MHz = data.freq
N_freqs = len(f_MHz)

for pp in range(2):
    if True:
        subplot(2, 2, 2*pp+1)
        for fb in range(N_freqs):
            if not rfi_mask[fb]:
                plot(beam_power_clipped[fb, pp], color='grey', alpha=0.6)
        plot(np.mean(beam_power_clipped[~rfi_mask, pp], axis=0), color='blue', lw=2)
        grid()
        ylabel('YY' if data['tiedbeam_locations']['pol'][pp].astype(str)=='S' else 'XX')
        if XLIM:
            xlim(xlims[0], xlims[1])
        if YLIM:
            ylim(ylims[0], ylims[1])
        if AV:
            axvline(x=time_index_frb_ds, ls='--', color='k')
        if pp==0:
            title('Clean frequencies')
            gca().set_xticklabels([])
        else:
            xlabel('time index (downsampled)')

        subplot(2, 2, 2*pp+2)
        for fb in range(N_freqs):
            if rfi_mask[fb]:
                plot(beam_power_clipped[fb, pp], color='grey', alpha=0.6)
        plot(np.mean(beam_power_clipped[rfi_mask, pp], axis=0), color='blue', lw=2)
        grid()

        if XLIM:
            xlim(xlims[0], xlims[1])
        if YLIM:
            ylim(ylims[0], ylims[1])
        if AV:
            axvline(x=time_index_frb_ds, ls='--', color='k')
        if pp==0:
            title('RFI frequencies')
            gca().set_xticklabels([])
        else:
            xlabel('time index (downsampled)')

suptitle('Cleaned,Beamformed and dedispersed data from %s\nDownsample factor = %i'
         %(acq_name, downsample_factor), y=1.05)
tight_layout()
```
